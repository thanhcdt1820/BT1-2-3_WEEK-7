```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
```

```python
url = 'Position_Salaries.csv'
data = pd.read_csv(url)
data.head()
```

```python
data.hist()
plt.show()
```

```python
sns.pairplot(data)
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Level']), y = data['Salary'])

print('Gia tri intercept', model.intercept_)
print('Gia tri coeffiient', model.coef_)
print('Do chinh xac', model.score(X = pd.DataFrame(data['Level']), y = data['Salary']))
```

```python
data.plot(kind = 'scatter', x='Level', y='Salary', figsize=(9,9), color='black', xlim=(0,10))
y_pred = model.predict(X = pd.DataFrame(data['Level']))
plt.plot(data['Level'], y_pred, color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Level'],data['Level']**2]).T
model.fit(X,y = data['Salary'])
print('Gia tri intercept', model.intercept_)
print('Gia tri coeffiient', model.coef_)
print('Do chinh xac', model.score(X,y = data['Salary']))
```

```python
range = np.arange(0,11,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)

data.plot(kind = 'scatter', x='Level', y='Salary', figsize=(9,9), color='black', xlim=(0,11))
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Level'],data['Level']**2,data['Level']**3,data['Level']**4,data['Level']**5,data['Level']**6]).T
model.fit(X,y = data['Salary'])
print('Gia tri intercept', model.intercept_)
print('Gia tri coeffiient', model.coef_)
print('Do chinh xac', model.score(X,y = data['Salary']))
```

```python
range = np.arange(0,11,0.1)
X_range = pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6]).T
y_pred = model.predict(X=X_range)

data.plot(kind = 'scatter', x='Level', y='Salary', figsize=(9,9), color='black', xlim=(0,11))
plt.plot(range,y_pred,color='blue')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
```

```python
url = 'passenger.csv'
data1 = pd.read_csv(url)
data1
```

```python
data1.hist()
plt.show()
```

```python
sns.pairplot(data1)
```

```python
data1.corr()
```

```python
sns.heatmap(data1.corr(), annot=True)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data1['Miles(X)']), y = data1['Dollars (Y)'])

print('Gia tri intercept', model.intercept_)
print('Gia tri coeffiient', model.coef_)
print('Do chinh xac', model.score(X = pd.DataFrame(data1['Miles(X)']), y = data1['Dollars (Y)']))
```

```python
data1.plot(kind = 'scatter', x='Miles(X)', y='Dollars (Y)', figsize=(9,9), color='Green', xlim=(0,6000))
y_pred = model.predict(X = pd.DataFrame(data1['Miles(X)']))
plt.plot(data1['Miles(X)'], y_pred, color='Pink')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data1['Miles(X)'],data1['Miles(X)']**2]).T
model.fit(X,y = data1['Dollars (Y)'])
print('Gia tri intercept', model.intercept_)
print('Gia tri coeffiient', model.coef_)
print('Do chinh xac', model.score(X,y = data1['Dollars (Y)']))
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data1['Miles(X)'],data1['Miles(X)']**2,data1['Miles(X)']**3,data1['Miles(X)']**4,data1['Miles(X)']**5]).T
model.fit(X,y = data1['Dollars (Y)'])
print('Gia tri intercept', model.intercept_)
print('Gia tri coeffiient', model.coef_)
print('Do chinh xac', model.score(X,y = data1['Dollars (Y)']))
```

```python
range = np.arange(0,6000,10)
X_range = pd.DataFrame([range,range**2,range**3,range**4,range**5]).T
y_pred = model.predict(X=X_range)

data1.plot(kind = 'scatter', x='Miles(X)', y='Dollars (Y)', figsize=(9,9), color='Green', xlim=(0,6000))
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
model.fit(X = data1.loc[:,['Miles(X)','Passenger']], y=data1['Dollars (Y)'])
print('Intercept la:', model.intercept_)
print('Coeffiient la:', model.coef_)
print('Do chinh xac', model.score(X = data1.loc[:,['Miles(X)','Passenger']], y=data1['Dollars (Y)']))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np 
```

```python
url = 'data_crab.csv'
data2 = pd.read_csv(url)
data2
```

```python
data2.hist()
plt.show()
```

```python
sns.pairplot(data2)
```

```python
data2.corr()
```

```python
sns.heatmap(data2.corr(), annot=True)
```

```python
data2.plot(kind = 'scatter',x = 'FL', y = 'BD', figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data2['FL']), y = data2['BD'])

print('Gia tri intercept', model.intercept_)
print('Gia tri coeffiient', model.coef_)
print('Do chinh xac', model.score(X = pd.DataFrame(data2['FL']), y = data2['BD']))
```

```python
data2.plot(kind = 'scatter', x='FL', y='BD', figsize=(9,9), color='Green', xlim=(0,55))
y_pred = model.predict(X = pd.DataFrame(data2['FL']))
plt.plot(data2['FL'], y_pred, color='Pink')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data2['FL'],data2['FL']**2,data2['FL']**3,data2['FL']**4]).T
model.fit(X,y = data2['BD'])
print('Gia tri intercept', model.intercept_)
print('Gia tri coeffiient', model.coef_)
print('Do chinh xac', model.score(X,y = data2['BD']))
```

```python
range = np.arange(0,22,0.5)
X_range = pd.DataFrame([range,range**2,range**3,range**4]).T
y_pred = model.predict(X=X_range)
data2.plot(kind = 'scatter', x='FL', y='BD', figsize=(9,9), color='red', xlim=(0,55))
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
model.fit(X = data2.loc[:,['FL','CL']], y=data2['BD'])
print('Intercept la:', model.intercept_)
print('Coeffiient la:', model.coef_)
print('Do chinh xac', model.score(X = data2.loc[:,['FL','CL']], y=data2['BD']))
```
